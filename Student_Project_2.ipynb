```python
X = em.drop(['Email No.', 'Prediction'], axis = 'columns')
y = em['Prediction']
```

```python
## Scikit-learn
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=47, stratify=y)
```

```python

# Buat objek model Linear Regression
model = LinearRegression()

# Latih model dengan data training
model.fit(X_train, y_train)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_model = LogisticRegression(max_iter=1000)
```

```python
logistic_model.fit(X_train, y_train)
```

```python
pred_1_LM = logistic_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print("\0.25[1m" + "LOGISTIC REGRESSION RESULT : " + "\030[0m")
print(classification_report(pred_1_LM, y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree_model = DecisionTreeClassifier()
dtree_model.fit(X_train, y_train)
pred_3_DC = dtree_model.predict(X_test)
print("\025[1m" + "DECISION TREE RESULT : " + "\025[0m")
print(classification_report(pred_3_DC, y_test))
```

```python
from sklearn.svm import SVC
```

```python
svm_model = SVC()
svm_model.fit(X_train, y_train)
pred_4_SVC = svm_model.predict(X_test)
print("\025[1m" + "SVM RESULT : " + "\025[0m")
print(classification_report(pred_4_SVC, y_test))
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
NB_model = GaussianNB()
NB_model.fit(X_train, y_train)
pred_5_NB = NB_model.predict(X_test)
print("\030[1m" + "NAIVE BAYES RESULT : "  + "\030[0m")
print(classification_report(pred_5_NB, y_test))
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
MPL_model = MLPClassifier()
MPL_model.fit(X_train, y_train)
pred_6_MPL = MPL_model.predict(X_test)
print("\025[1m" + "NEURAL NETWORK RESULT : " + "\025[0m")
print(classification_report(pred_6_MPL, y_test))
```
